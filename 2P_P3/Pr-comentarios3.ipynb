#limpieza y mantenimiento a dichos datos archivos que cuentan con diversos tipos de datos relacionados a los cereales.

#Paso 1: El primer paso que se realizó fue importar pandas y leer el archivo que previamente nos habían proporcionado 
#"cereal.csv".

```python
from pandas import read_csv, DataFrame, merge, concat
```

#Paso 2: Posterior a eso creamos un DataFrame al que se le asignan los primeros 20 registros de cereales.

```python
#Punto 1
cereal = read_csv('cereals.csv')

scereal = DataFrame(cereal, 
                    index = range(20), 
                    columns = ['Name', 'Manuf', 'Type', 'Calories', 'Protein', 'Sodium', 'Potass', 'Vitamins' ])
scereal
```

#Paso 3: Con la funciones "mean" obtenemos la media, "std" la desviación estandar y "sqrt" la raíz cuadrada del Sodio y el 
#Calcio que contienen los cereales.

```python
#Punto 2
import numpy as np


df1 = DataFrame(round(scereal[['Calories', 'Sodium']].aggregate([np.mean, np.std]),2)).T

#df2 = DataFrame(round(np.sqrt((scereal[['Calories', 'Sodium']].apply(sum))), 2), columns = ["sqrt"])
#df1.merge(df2, left_index=True, right_index=True)

df1['sqrt'] = DataFrame(round(np.sqrt((scereal[['Calories', 'Sodium']].apply(sum))), 2), columns = ["sqrt"])
df1
```

#Paso 4: Con la función "replace" sustituimos los valores por la media de Sodio.

```python
# Punto 3
cereal.replace(0, round(cereal['Sodium'].mean(), 2))
```

#Paso 5: Utilizamos la función "dropna" para eliminar los valores nulos.

```python
#Punto 4
sncereal =  cereal.dropna()
sncereal
```

#Paso 6: Con la función "describe" obtenemos la media, la desviación estándar, los datos mínimos, los datos máximos, etc. 

```python
#Punto 5

sncereal.describe()
```

#Paso 7: Posterior a eso en el siguiente paso hacemos uso del operador lógico "or" | y filtramos los datos que en Manuf 
#contengan una K o en Potass sean mayores o iguales a 150.

```python
#Punto 6

cereal.loc[(cereal['Manuf'].str.contains("K")) & (cereal['Potass'] >= 150)]
```

#Paso 8: Con la función "groupby" sumamos los campos de Calories y los ordenanamos con "ascending = False"Falseo que hará que 
#los datos se ordenen de forma descendente.

```python
#Punto 7
DataFrame(scereal.groupby(['Name', 'Type'],as_index=True)['Calories'].sum()).sort_values('Calories', ascending = False)
```

#Paso 9: Con las funciones que previamente ya habíamos utilizado "mean", "std" y "sqrt" sacamos la media, la desviación estandar 
#y la raíz cuadrada de los campos de tipo cuantitativo del DataFrame "scereal". 

```python
#Punto 8
sqrts = DataFrame(round(np.sqrt((scereal[['Calories', 'Protein', 'Sodium', 'Potass', 'Vitamins']].apply(sum))), 2), columns = ["sqrt"])
mean_n_std = scereal.apply(['mean', 'std']).T
mean_n_std.merge(sqrts, left_index=True, right_index=True)
```

#Paso 10: Posterior a eso unificamos los archivos en un mismo DataFrame que llamamos "supercereal".

```python
#Punto 9
#supercereal = merge(cereal, read_csv('cereals2.csv'), left_on="Name", right_on="Name")
#supercereal

supercereal = concat([cereal, read_csv('cereals2.csv')])
supercereal
```

#Paso 11: Y para finalizar de igual forma que en el paso 9 con la funciones que previamente ya habiamos utilizado "mean", "std"
#y "sqrt" sacamos la media, la desviación estándar y la raíz cuadrada de los datos cuantitativos del DataFrame supercereal.

```python
#Punto 10
sqrts2 = DataFrame(round(np.sqrt((supercereal[['Calories', 'Protein', 'Sodium', 'Potass', 'Vitamins']].apply(sum))), 2), columns = ["sqrt"])
mean_n_std2 = supercereal.apply(['mean', 'std']).T
mean_n_std2.merge(sqrts2, left_index=True, right_index=True)
```
